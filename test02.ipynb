```python
# 1.编写 Python 程序输出一个具有如下风格效果的文本，用作文本进度条样式，部分代码如下，填写空格处。
#  10%@==
#  20%@====
# 100%@====================
 
# 前三个数字，右对齐；后面字符，左对齐
 
# 文本中左侧一段输出 N 的值，右侧一段根据 N 的值输出等号，中间用 @ 分隔，等号个数为 N 与 5 的整除商的值，
# 例如，当 N 等于 10 时，输出 2 个等号。
```

```python
n = eval(input())

print("{:>3}%@{}".format(n, "="*(n//5)))
```

```python
# 3.使用程序计算整数 N 到整数 N+100 之间所有奇数的数值和，不包含 N+100，并将结果输出。
# 整数 N 由用户给出，代码片段如下，补全代码。不判断输入异常。
```

```python
n = input()
s = 0
for i in range(eval(n), eval(n)+100): 
    if i % 2 == 1:
        s += i     
print(s)
```

```python
# 5.经常会有要求用户输入整数的计算需求，但用户未必一定输入整数。为了提高用户体验，编写 getInput() 函数处理这样的情况。
# 请补充如下代码，如果用户输入整数，则直接输出整数并退出，如果用户输入的不是整数，则要求用户重新输入，直至用户输入整数为止。
```

```python
def getInput():
    try:
        n = input("请输入整数：")
        while eval(n) != int(n):
            n = input("请输入整数：")
    except:
        return getInput()     
    return eval(n) 
print(getInput())
```

```python
# 问题1：请编写程序，对这个《天龙八部》文本中出现的汉字和标点符号进行统计，字符与出现次数之间用冒号:分隔，
#     输出保存到“天龙八部-汉字统计.txt”文件中，该文件要求采用 CSV 格式存储，参考格式如下（注意，不统计空格和回车字符）：
# 天:100, 龙:110, 八:109, 部:10
```

```python
with open("H:/二级/05操作题/2/天龙八部节选-网络版.txt", "r", encoding="utf-8") as f:
    file = f.readlines()

dic = {}
text = "".join(file)
for i in text:
    if(i != " " and i != "\n" and i != "\ufeff" and i != "\u3000"):
        dic[i] = dic.get(i, 0) + 1
    
# lis = []
# for key in dic:
#     lis.append("{}:{}".format(key, dic[key]))
lis = ["{}:{}".format(key, dic[key]) for key in dic]
print(lis)

with open("H:/二级/05操作题/2/天龙八部节选-汉字统计.txt", "w", encoding="utf-8") as f:
    f.write(",".join(lis))

print("ojbk")
```

```python
问题2：请编写程序，对《天龙八部》文本中出现的中文词语进行统计，采用 jieba 库分词，词语与出现次数之间用冒号:分隔，
    输出保存到“天龙八部-词语统计.txt”文件中。参考格式如下（注意，不统计任何标点符号）：
天龙:100, 八部:10
```

```python
import jieba
with open("H:/二级/05操作题/2/天龙八部节选-网络版.txt", "r", encoding="utf-8") as f:
    file = f.read()

words = jieba.lcut(file)
                   
dic = {}
for word in words:
    if(word != " " and word != "\n" and word != "\ufeff" and word != "\u3000"):
        dic[word] = dic.get(word, 0) + 1
        
lis = ["{}:{}".format(k, dic[k]) for k in dic]
# print(lis)

with open("H:/二级/05操作题/2/天龙八部节选-词语统计.txt", "w", encoding="utf-8") as f:
    f.write(",".join(lis))
print("ojbk")
```
